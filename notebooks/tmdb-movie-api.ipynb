# Movie API

```python
import os
import requests as r
from PIL import Image
from IPython.display import display
```

```python
bearer_token = os.environ['TMDB_API_KEY']
```

### Movie Discovery

```python
movie_discovery = r.get("https://api.themoviedb.org/3/discover/movie?release_date.gte=2019-05-15&release_date.lte=2019-05-16&language=pt-br&sort_by=release_date.asc&page=6", 
                        headers = {"Authorization": f"Bearer {bearer_token}"}).json()
```

```python
movie_results = movie_discovery.get('results')
```

```python
movie_results[0:2]
```

### Movie Details

```python
movie_details = r.get("https://api.themoviedb.org/3/movie/458897?append_to_response=credits,images,external_ids&language=pt-br", 
                      headers = {"Authorization": f"Bearer {bearer_token}"}).json()
```

```python
m_casting = movie_details.get('credits').get('cast')
m_genres = movie_details.get('genres')
m_images = movie_details.get('images')
m_external_ids = movie_details.get('external_ids')
```
