# Search API

```python
import os
import requests as r
from typing import List
from PIL import Image
from IPython.display import display
```

```python
bearer_token = os.environ['TMDB_API_KEY']
```

### Multi Search - tv :: The Good Doctor

```python
multi_search = r.get("https://api.themoviedb.org/3/search/multi?query=the+good+doctor+-++o+bom+doutor&language=pt-br", 
                     headers = {"Authorization": f"Bearer {bearer_token}"}).json()
```

```python
multi_search
```

### Multi Search - movie :: John Wick

```python
multi_search = r.get("https://api.themoviedb.org/3/search/multi?query=john+wick&language=pt-br", 
                     headers = {"Authorization": f"Bearer {bearer_token}"}).json()
```

```python
search_results = multi_search.get('results')
```

```python
sorted(search_results, reverse=True, key=lambda result: result.get('popularity'))    
```
