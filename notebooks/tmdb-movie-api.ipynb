# Movie API

```python
import os
import requests as r
from PIL import Image
from IPython.display import display
```

```python
bearer_token = os.environ['TMDB_API_KEY']
```

### Movie Discovery

```python
movie_discovery = r.get("https://api.themoviedb.org/3/discover/movie?release_date.gte=2019-05-15&release_date.lte=2019-05-16&sort_by=release_date.asc&page=6", 
                        headers = {"Authorization": f"Bearer {bearer_token}"}).json()
```

```python
movie_results = movie_discovery.get('results')
```

```python
movie_results[3:4]
```

### Movie Details

```python
movie_details = r.get("https://api.themoviedb.org/3/movie/458156?append_to_response=videos,images", 
                      headers = {"Authorization": f"Bearer {bearer_token}"}).json()
```

```python
videos = movie_details.get('videos')
images = movie_details.get('images')
```

```python
movie_details.get('release_date')
```

### Movie Casting

```python
movie_casting = r.get("https://api.themoviedb.org/3/movie/458156/credits", 
                      headers = {"Authorization": f"Bearer {bearer_token}"}).json()
```

```python
movie_casting_results = movie_casting.get('cast')
```

```python
character = movie_casting_results[0]
```

```python
character
```

### Movie Actor Images

```python
image_url = f"https://www.themoviedb.org/t/p/original/{character.get('profile_path')}"
```

```python
image = r.get(image_url, stream=True).raw
```

```python
display(Image.open(image))
```
